# Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Serial Test

```python
WT = np.loadtxt('/content/drive/MyDrive/WT_serial.txt')
```

```python
P = 2
Q = 4
imgsize = 512
WT = WT.reshape((P*Q + 2, 512, 512))
plt.figure(figsize = (15, 8))
count = 1

for n in range(P):
    for m in range(Q):
        plt.subplot(P, Q, count)
        plt.imshow(np.abs(WT[1+count]), cmap = 'Greys')
        plt.title('scale = {}, orientation = {}'.format(n, m*180/Q))
        count += 1
plt.suptitle('Directional Filters (Serial)')
```

```python
plt.figure(figsize = (9, 4))
plt.subplot(1, 2, 1)
plt.imshow(np.abs(WT[0]), cmap = 'Greys')
plt.title('Low Pass (Serial)')
plt.subplot(1, 2, 2)
plt.imshow(np.abs(WT[1]), cmap = 'Greys')
plt.title('High Pass (Serial)')
```
